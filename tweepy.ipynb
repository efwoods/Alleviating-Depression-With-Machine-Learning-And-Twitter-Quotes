```python
import pandas as pd
import configparser
import tweepy
import requests
import random

# read config
config = configparser.ConfigParser()
config.read('./code/config/config.ini')

api_key = config['twitter']['api_key']
api_key_secret = config['twitter']['api_key_secret']

access_token = config['twitter']['access_token']
access_token_secret = config['twitter']['access_token_secret']

# authenticate
auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
public_tweets = api.home_timeline()
print(public_tweets)
```

```python
evanWoods = api.search_full_archive(screen_name="EvanWoods")
```

```python
evanWoods.screen_name
```

```python
new_tweets = api.user_timeline(screen_name = evanWoods.screen_name,count=200, tweet_mode="extended")
```

```python
tweets = [[tweet.full_text] for tweet in new_tweets]
```

```python
len(tweets)
```

```python
for tweet in tweets:
    print(tweet)
```

```python
def parse_fav_quote():
    url = "https://efwoods.github.io/EvanWoodsFavoriteQuotes/quotesTwitterDB.json"
    fav_quote = requests.request("GET", url).json()
    quote = random.randint(0, len(fav_quote["quotes"]))
    return fav_quote["quotes"][quote]
```

```python
test = parse_fav_quote()
```

```python
for tweet in tweets:
    if test in tweet:
        print(test)
    # else:
    #     print("test not in tweets")
```
